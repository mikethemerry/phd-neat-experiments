```python
def export_graph(source, experiment_name, iteration, params):
    epochs = params[-3:-1]
    file_name = "{}_it-{}_epo-{}".format(experiment_name, iteration, epochs)
    mygraph = graphviz.Source(source, format="png")
    mygraph.render(file_name, format="png")
```

```python
resultsDB.data.loc[resultsDB.data['measurementType'] == "best_genome_map"].apply(lambda x: export_graph(x.value, x.experimentName, x.iteration, x.params), axis=1)
    
```
